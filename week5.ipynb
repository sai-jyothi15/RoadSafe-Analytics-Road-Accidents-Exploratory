```python
#STEP 1: Import Libraries & Load Required Columns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['Start_Lat','Start_Lng','Severity','State','City']
df = pd.read_csv("../cleaned_us_accidents.csv", usecols=cols)
df = df.dropna()
df.head()
```

```python
#STEP 2: Scatter Plot of Accident Points
plt.figure(figsize=(6,6))
plt.scatter(df['Start_Lat'][0:5000], df['Start_Lng'][0:5000], s=2, alpha=0.3)
plt.title("Accident Hotspots (Sample Data)")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
```

```python
#STEP 3: Top 5 Most Accident-Prone States
top_states = df['State'].value_counts().head(5)
print("Top 5 Accident States:")
print(top_states)

top_states.plot(kind='bar', figsize=(7,4), color='orange')
plt.title("Top 5 Accident-Prone States")
plt.xlabel("State")
plt.ylabel("Accident Count")
plt.show()
```

```python
#STEP 4: Top 5 Most Accident-Prone Cities
top_cities = df['City'].value_counts().head(5)
print("Top 5 Cities with Highest Accidents")

top_cities.plot(kind='bar', figsize=(8,4), color='purple')
plt.title("Top 5 Accident Cities")
plt.xlabel("City")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```
