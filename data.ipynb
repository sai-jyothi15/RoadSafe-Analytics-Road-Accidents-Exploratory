```python
import pandas as pd

df = pd.read_csv("cleaned_us_accidents.csv", low_memory=False)

print(df.shape)
print(df.columns)
```

```python
# Select important columns for dashboard
cols = [
    'Severity',
    'Start_Time',
    'City',
    'State',
    'Weather_Condition',
    'Visibility(mi)'
]

df_dash = df[cols]

# Remove missing values
df_dash = df_dash.dropna()

print("After cleaning:", df_dash.shape)

# Take sample for Streamlit (performance)
df_sample = df_dash.sample(20000, random_state=42)

# Save dashboard data
df_sample.to_csv("dashboard_sample.csv", index=False)

print("dashboard_sample.csv created successfully")
df_sample.head()
```
