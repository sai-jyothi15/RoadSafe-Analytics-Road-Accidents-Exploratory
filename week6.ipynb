```python
#QUESTION 1: What time of day has the most accidents?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_time = pd.read_csv("../cleaned_us_accidents.csv", usecols=['Start_Time'])
df_time['Start_Time'] = pd.to_datetime(df_time['Start_Time'], errors='coerce')
df_time['Hour'] = df_time['Start_Time'].dt.hour

sns.countplot(x=df_time['Hour'], palette="viridis")
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Accident Count")
plt.show()

df_time['Hour'].value_counts().head()
```

```python
#QUESTION 2: Are accidents more severe during rain or fog?
cols2 = ['Severity','Weather_Condition']
df_w = pd.read_csv("../cleaned_us_accidents.csv", usecols=cols2).dropna()

weather_focus = df_w[df_w['Weather_Condition'].isin(['Rain','Fog','Clear'])]

sns.boxplot(x='Weather_Condition', y='Severity', data=weather_focus)
plt.title("Severity during Rain, Fog & Clear Weather")
plt.show()
```

```python
#QUESTION 3: Is visibility correlated with severity?
cols3 = ['Severity','Visibility(mi)']
df_vis = pd.read_csv("../cleaned_us_accidents.csv", usecols=cols3).dropna()

correlation = df_vis.corr()
print(correlation)

sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation between Visibility & Severity")
plt.show()
```
